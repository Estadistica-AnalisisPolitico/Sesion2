```python
link="https://www.theglobaleconomy.com/rankings/wb_government_effectiveness/"

import pandas as pd
import requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"
}
# Make the request with the custom headers
response = requests.get(link, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using pandas
    datas = pd.read_html(response.content)
else:
    print(f"Error: Status code {response.status_code}")
```

```python
effec=datas[0]
effec.info()
```

```python
effec=effec.iloc[:,:2]
effec.rename(columns={"Government effectiveness, 2023":"effectiveness"},inplace=True)
```

```python
link2="https://en.wikipedia.org/wiki/The_Economist_Democracy_Index"

data2=pd.read_html(link2,match="Year 2024")
```

```python
demo=data2[2]
```

```python
demo.head()
```

```python
demo.columns=[x[0] for x in demo.columns]
```

```python
demo.head()
```

```python
bye=demo.columns[[1,5]]
demo.drop(columns=bye,inplace=True)
demo
```

```python
demo.info()
```

```python
toNumAll=lambda x:pd.to_numeric(x,errors="coerce")
demo[demo.columns[3:]]=demo.iloc[:,3:].apply(toNumAll)
```

```python
demo.info()
```

```python
demo[demo.iloc[:,3].isna()]
```

```python
demo=demo[~demo.iloc[:,3].isna()]
demo.reset_index(drop=True,inplace=True)
```

```python
demo.columns.to_list()
```

```python
import re
demo.columns=[re.sub("\xad","",x) for x in demo.columns.to_list()]
```
